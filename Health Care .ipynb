```python
# data prep - to train the model on training data
#  find the accuracy of prediction of the trained model on the test data

# 70 % of the data as training data and 30 % as the test data

from sklearn.model_selection import train_test_split
```

```python
# train_test_split(X-independent var, y- dependent variable,test_size=0.3,random_state=40)
```

```python
df.columns
```

```python
# ind. vars

X=df.drop('Outcome',axis=1)
```

```python
X.head()
```

```python
df.head()
```

```python
y=df['Outcome']
```

```python
y[:5]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# train the model on the training dataset

# import the logistic regression class


from sklearn.linear_model import LogisticRegression
```

```python
# create an instance of the Logistic regression class

log_regression=LogisticRegression(max_iter=200)
```

```python
# log_regression
# fit the logistic regression class on the training data
log_model=log_regression.fit(X_train,y_train)
```

```python
# use the trained model to predict on the test data
y_prediction=log_model.predict(X_test)
```

```python
y_prediction
```

```python
y_test.values
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_prediction)
```

```python
print(classification_report(y_test,y_prediction))
```

```python
# KNN-

from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model=knn.fit(X_train,y_train)
```

```python
y_pred=knn_model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# random forest

from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest=RandomForestClassifier()
```

```python
rand_model=random_forest.fit(X_train,y_train)
```

```python
y_pred=rand_model.predict(X_test)
```

```python
y_pred
```

```python
print(classification_report(y_test,y_pred))
```
